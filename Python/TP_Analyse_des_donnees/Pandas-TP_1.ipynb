## Étape 2. Attribuez-le à une variable appelée drinks.

```python
url = "https://drive.google.com/file/d/1e4gZTgB3sGFqelPKv0ppEkpn4cCIS_jz/view?usp=sharing"
url ='https://drive.google.com/uc?id=' + url.split('/')[-2]
#url = "data.tsv"
drinks=pd.read_csv(url, sep=',')
drinks
```

## Étape 3. Quel continent boit plus de l'alcool en moyenne?

```python
#print(drinks.total_litres_of_pure_alcohol.groupby(drinks.continent).max())
print(drinks.groupby(drinks.continent).total_litres_of_pure_alcohol.mean().sort_values(ascending=False))
```


## Étape 4. Pour chaque continent, afficher les statistiques de la consommation de l'alcool.

```python
print(drinks.groupby(drinks.continent).describe())
```

## Étape 5. Afficher la consommation moyenne d'alcool par continent pour chaque colonne

```python
print(drinks.groupby(drinks.continent).mean())
```

## Étape 6. Afficher la consommation médiane d'alcool par continent pour chaque colonne

```python
print(drinks.groupby(drinks.continent).median())
```
